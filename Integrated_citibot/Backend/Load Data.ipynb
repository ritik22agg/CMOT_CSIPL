```python
table = "test2"
# create table 
create = "CREATE TABLE " + table + "("
cnt = 0
for cols in data.columns:
    create += (cols.lower())
    if(data[cols].dtype == 'int64'):
        create += ' int'
    else:
        create += ' varchar(30)'
    
    if cnt < len(data.columns)-1 :
        create += ','
    cnt = cnt + 1
    
create += ');'
print(create)  
```
